```python
import pandas as pd
```

```python
col_names = [
    "Surgery",
    "Age",
    "Rectal temp",
    "Pulse",
    "Respiratory rate",
    "Temperature of extremites",
    "Pain",
    "Outcome",
]
hd = pd.read_csv(
    "horse_data.csv",
    header=None,
    usecols=[0, 1, 3, 4, 5, 6, 10, 22],
    names=col_names,
    na_values=["?"],
)
hd.head()
```

```python
surgery_outlier = hd[(hd["Surgery"] < 1) | (hd["Surgery"] > 2)]

age_outlier = hd[(hd["Age"] < 1) | (hd["Age"] > 2)]
hd["Age"] = hd["Age"].replace(9, 1)

hd["Rectal temp"] = hd["Rectal temp"].fillna(37.8)

hd.head(10)
```

```python
# find outlier in rectal temp
def find_outliers(data):
    iqr = data.quantile(0.75) - data.quantile(0.25)
    lower_bound = data.quantile(0.25) - (iqr * 1.5)
    upper_bound = data.quantile(0.75) + (iqr * 1.5)
    return data[(data < lower_bound) | (data > upper_bound)]


rectal_temp_outlier = find_outliers(hd["Rectal temp"])
rectal_temp_outlier = hd[hd["Rectal temp"].isin(rectal_temp_outlier)]
rectal_temp_outlier

hd.drop(rectal_temp_outlier.index, inplace=True)

hd.head()
```

```python
# find outliers in pulse

pulse_na = hd[hd["Pulse"].isna()]
hd.drop(pulse_na.index, inplace=True)

pulse_outliers = find_outliers(hd["Pulse"])
pulse_outliers
```

```python
hd["Respiratory rate"] = hd["Respiratory rate"].fillna(
    9
)  # cuz - usefulness is doubtful due to the great fluctuations
```

```python
temp_index = hd[hd["Temperature of extremites"].isna()]
hd.drop(temp_index.index, inplace=True)
hd.head(10)
```

```python
# - hot extremities should correlate with an elevated rectal temp.

hd.drop(
    hd[(hd["Temperature of extremites"] == 2) & (hd["Rectal temp"] <= 37.8)].index,
    inplace=True,
)
hd.head(10)
```
