```python
import pandas as pd
```

```python
col_names = [
    "Surgery",
    "Age",
    "Rectal temp",
    "Pulse",
    "Respiratory rate",
    "Temperature of extremites",
    "Pain",
    "Outcome",
]
hd = pd.read_csv(
    "horse_data.csv",
    header=None,
    usecols=[0, 1, 3, 4, 5, 6, 10, 22],
    names=col_names,
    na_values=["?"],
)
hd.head()
```

```python
# find outliners in Series func


def find_outliers(data):
    iqr = data.quantile(0.75) - data.quantile(0.25)
    lower_bound = data.quantile(0.25) - (iqr * 1.5)
    upper_bound = data.quantile(0.75) + (iqr * 1.5)
    return data[(data < lower_bound) | (data > upper_bound)]
```

```python
# search of losed data and outliners in surgery col

surgery_mode = hd["Surgery"].mode()

hd["Surgery"] = hd["Surgery"].fillna(
    value=surgery_mode[0]
)  # fill empty values with moode

surgery_outlier = hd[(hd["Surgery"] < 1) | (hd["Surgery"] > 2)]
print(surgery_outlier)
hd.describe()
```

```python
# search of losed data and outliners in Age

age_outlier = hd[(hd["Age"] < 1) | (hd["Age"] > 2)]  # 9 was found

hd["Age"] = hd["Age"].replace(9, 2)

hd.describe()
```

```python
# search of losed data and outliners in Rectal temp

rectal_na = hd[hd["Rectal temp"].isna()]
# print(rectal_na)
rectal_outlier = find_outliers(rectal_na["Rectal temp"])  # not found
print("rectal_outlier", rectal_outlier)

hd["Rectal temp"] = hd["Rectal temp"].fillna(hd["Rectal temp"].mean())

hd.describe()  # after replacing nan rectal temp with mean, mean is not changed
```

```python
# search of losed data and outliners in Pulse

pulse_outliers = find_outliers(hd["Pulse"])  # found
# print(pulse_outliers)

pulse_na = hd[hd["Pulse"].isna()]

# changing pulse NaN column into mode
hd["Pulse"] = hd["Pulse"].fillna(hd["Pulse"].mode()[0])

hd.describe()  # pulse mean slightly changed
```

```python
# search of losed data and outliners in Respiratory rate

respiratory_rate_outliers = find_outliers(hd["Respiratory rate"])
# ignore? due to - usefulness is doubtful due to the great fluctuations

respiratory_rate_na = hd[hd["Respiratory rate"].isna()]

hd["Respiratory rate"] = hd["Respiratory rate"].fillna(hd["Respiratory rate"].mode()[0])

hd.describe()
```

```python
# search of losed data and outliners in Temperature of extremites

temp_extr_outliers = find_outliers(hd["Temperature of extremites"])  # none
temp_extr_na = hd[hd["Temperature of extremites"].isna()]
# print(temp_extr_na)

hd["Temperature of extremites"].fillna(
    hd["Temperature of extremites"].mode()[0], inplace=True
)


hd.describe()
```

```python
# search of losed data and outliners in Pain

pain_outliers = hd[(hd["Pain"] < 1) | (hd["Pain"] > 5)]

hd["Pain"].fillna(hd["Pain"].mode()[0], inplace=True)

hd.describe()
```

```python
# search of losed data and outliners in Outcome

outcome_outliers = hd[(hd["Outcome"] < 1) | (hd["Outcome"] > 3)]

hd["Outcome"].fillna(hd["Outcome"].mode()[0], inplace=True)
hd.describe()
```
